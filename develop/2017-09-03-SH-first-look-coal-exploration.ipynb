```python
% matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
sns.set()
```

```python
# Read in our data and promote the 3rd row to the header row and index on the MSHA ID column
df = pd.read_excel("../data/coalpublic2013.xls", header=2, index_col='MSHA ID')
df.head()
```

```python
# Listing unique values within column
df['Company Type'].unique()
```

```python
# Replacing error values within column observations
df['Company Type'].replace(to_replace='Indepedent Producer Operator',
                           value='Independent Producer Operator',
                           inplace=True)

df['Company Type'].head()
```

```python
# Rename all columns to replace spaces with underscores (to make referencing easier)
df.rename(columns=lambda x: x.replace(" ", "_"), inplace=True)
df.head()
```

```python
len(df)
```

```python
# Scatter plot of two variables, note that now that we have underscores in
# variable names, we can use the df.column name index notation rather than 
# df['Column name']
plt.scatter(df.Average_Employees, df.Labor_Hours)
plt.xlabel('Number of employees')
plt.ylabel('Total Hours Worked')
```

```python
# Seaborn regression plot and save the output plot to the figures folder
sns.regplot(df.Average_Employees, df.Labor_Hours)
plt.savefig("../figures/2017-09-03" + "-employees_vs_hours.png")
```

```python
# List columns
for column in df.columns:
    print(column)
```

```python
plt.scatter(df.Labor_Hours, df['Production_(short_tons)'])
```

```python
df['Production_(short_tons)'].hist()
```

We see a very uneven distribution and as such may want to transform this variable.

```python
# Passing df['Production_(short_tons)'] which gives a list of Booleans to the the dataframe
# is effectively saying return the dataframe where Production_(short_tons) is 0
df[df['Production_(short_tons)'] == 0]
```

```python
# Values that are not zero
# We are removing data here!
df_nonzero = df[df['Production_(short_tons)'] > 0]
len(df_nonzero)
```

```python
df_nonzero['Production_(short_tons)'].hist()
```

We still have a skewed distribution, so we consider a log transformation

```python
# Log transforming our target variable
df_nonzero['log_production'] = np.log(df_nonzero['Production_(short_tons)'])
df_nonzero.head()
```

```python
df_nonzero['log_production'].hist()
```

From our results, it seems that production follows a log-normal distribution.
